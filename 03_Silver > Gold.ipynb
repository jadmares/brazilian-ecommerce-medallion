```python
%sql
CREATE OR REPLACE TABLE `brazillian-e-commerce`.01_gold.sales_analytics
AS

SELECT

    -- Pedido
    o.order_id,
    o.order_status,
    o.purchase_ts,
    o.approved_ts,
    o.delivered_ts,
    o.estimated_dt,

    -- Cliente
    c.customer_id,
    c.customer_unique_id,
    c.customer_city,
    c.customer_state,

    -- Item
    oi.order_item_id,
    oi.product_id,
    oi.seller_id,
   -- oi.shipping_limit_ts,
    oi.price,
    oi.freight_value,

    -- Producto
    p.product_category_name,
    p.product_name_length,
    p.product_description_length,
    p.product_photos_qty,
    p.product_weight_g,
    p.product_length_cm,
    p.product_height_cm,
    p.product_width_cm,

    -- Seller
    s.seller_city,
    s.seller_state,

    -- Pagos
    op.payment_type,
    op.payment_installments,
    op.payment_value,

    -- Reviews
    r.review_score,
    r.review_creation_date,
    r.review_answer_ts

FROM `brazillian-e-commerce`.02_silver.orders o

INNER JOIN `brazillian-e-commerce`.02_silver.orderItems oi
    ON o.order_id = oi.order_id

LEFT JOIN `brazillian-e-commerce`.02_silver.customers c
    ON o.customer_id = c.customer_id

LEFT JOIN `brazillian-e-commerce`.02_silver.products p
    ON oi.product_id = p.product_id

LEFT JOIN `brazillian-e-commerce`.02_silver.sellers s
    ON oi.seller_id = s.seller_id

LEFT JOIN `brazillian-e-commerce`.02_silver.orderPayments op
    ON o.order_id = op.order_id

LEFT JOIN `brazillian-e-commerce`.02_silver.orderReviews r
    ON o.order_id = r.order_id;
```
