```sql
-- Crear la tabla si no existe, pero vacía
CREATE TABLE IF NOT EXISTS `brazillian-e-commerce`.03_bronze.customers
USING DELTA
AS
SELECT *
FROM `brazillian-e-commerce`.landing.olist_customers_dataset
WHERE 1 = 0;

-- Carga inicial o incremental
MERGE INTO `brazillian-e-commerce`.03_bronze.customers AS target
USING `brazillian-e-commerce`.landing.olist_customers_dataset AS source
ON target.customer_id = source.customer_id

WHEN MATCHED THEN UPDATE SET * 
WHEN NOT MATCHED THEN INSERT *
; 
```

```sql
CREATE TABLE IF NOT EXISTS `brazillian-e-commerce`.03_bronze.orders
USING DELTA
AS 
SELECT *
FROM `brazillian-e-commerce`.landing.olist_orders_dataset
WHERE 1=0;

MERGE INTO `brazillian-e-commerce`.03_bronze.orders AS target
USING `brazillian-e-commerce`.landing.olist_orders_dataset AS source
ON target.order_id = source.order_id
AND target.customer_id = source.customer_id

WHEN MATCHED THEN UPDATE SET *
WHEN NOT MATCHED THEN INSERT *
```

```sql
CREATE TABLE IF NOT EXISTS `brazillian-e-commerce`.03_bronze.orderItems
USING DELTA
AS
SELECT *
FROM `brazillian-e-commerce`.landing.olist_order_items_dataset
WHERE 1 = 0;

MERGE INTO `brazillian-e-commerce`.03_bronze.orderItems AS target
USING `brazillian-e-commerce`.landing.olist_order_items_dataset AS source
ON target.order_id = source.order_id
AND target.order_item_id = source.order_item_id

WHEN MATCHED THEN
    UPDATE SET *

WHEN NOT MATCHED THEN
    INSERT *;
```

```sql
CREATE OR REPLACE TABLE `brazillian-e-commerce`.03_bronze.orderPayments AS
SELECT *
FROM `brazillian-e-commerce`.landing.olist_order_payments_dataset;

SELECT
    (SELECT COUNT(*) FROM `brazillian-e-commerce`.landing.olist_order_payments_dataset) AS landing_count,
    (SELECT COUNT(*) FROM `brazillian-e-commerce`.03_bronze.orderPayments) AS bronze_count;
```
